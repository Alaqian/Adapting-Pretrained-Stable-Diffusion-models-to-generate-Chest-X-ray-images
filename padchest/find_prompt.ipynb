```python
import os
import cv2
import matplotlib.pyplot as plt
import shutil

txt_dir = "/home/aaz7118_nyu_edu/padchest/filtered_labels_RGB"
img_dir = "/home/aaz7118_nyu_edu/padchest/filtered_labels_RGB"
search_term = "pacemaker"
dest = dest = "/home/aaz7118_nyu_edu/tests/img325+diagnosis/pacemaker"
count = 0
# for all .txt files in txt_dir, find the ones with search_term and display the .png image in img_dir with the same name as the .txt file
for file in os.listdir(txt_dir):
    if file.endswith(".txt"):
        with open(os.path.join(txt_dir, file), "r") as f:   
            content = f.read()
            if search_term in content:
                count += 1
                # get the corresponding image file name
                img_file = os.path.join(img_dir, os.path.splitext(file)[0] + ".png")
                # display the image
                img = cv2.imread(img_file)
                # make the background black colored
                fig = plt.gcf()
                fig.patch.set_facecolor('black')
                plt.imshow(img)
                # make the title green
                plt.suptitle(str(count)+" \n" + str(file) + " \n" +"\n".join(content[i:i+80] for i in range(0, len(content), 80)), color="green")
                plt.axis("off")
                plt.tight_layout()
                print(count, os.path.splitext(file)[0] + ".png")
                plt.show()
                shutil.copy(os.path.join(txt_dir, file), dest)
                shutil.copy(os.path.join(img_dir, img_file), dest)
```
